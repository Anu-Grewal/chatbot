```python
from bs4 import BeautifulSoup
import pandas as pd

# Path to the HTML file
html_file_path = '/content/pib.html'

# Read the HTML content from the file
with open(html_file_path, 'r', encoding='utf-8') as file:
    html_content = file.read()
# Parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Find all the list items (li) containing the required information
articles = []
for section in soup.find_all('li'):
    # Extract the heading
    heading = section.find('h3', class_='font104')
    if heading:
        heading_text = heading.get_text(strip=True)
        # Find all the list items (li) within the section
        for li in section.find_all('li'):
            # Extract the content from the a tag
            a_tag = li.find('a')
            if a_tag:
                content = a_tag.get_text(strip=True)
                href = a_tag['href']
                # Extract the date from the span tag
                date_tag = li.find('span', class_='publishdatesmall')
                if date_tag:
                    date_of_publication = date_tag.get_text(strip=True).replace('Posted On: ', '')
                else:
                    date_of_publication = None
                # Append the extracted information to the list
                articles.append({
                    'Date of Publication': date_of_publication,
                    'Headline Article': heading_text,
                    'Content': content,
                    'Article Link': 'https://www.pib.gov.in' + href
                })
# Convert the list of dictionaries to a DataFrame
df = pd.DataFrame(articles)

# Save the DataFrame to an Excel file
df.to_excel('pib_articles.xlsx', index=False)

print("Data has been written to 'pib_articles.xlsx'")
```
